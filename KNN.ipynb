```python
# Euclidean Distance is Calculated

def euclidean_dist(test,train):
    dist = 0.0
    for i in range(len(test)):
        dif = train[i] - test[i]
        sq_dif = np.square(dif)
        dist += sq_dif
        r = np.sqrt(dist)
    return r
```

```python
def KNN(K):
    result = []
    for i in range(len(X_test)):
        distances = []
        for j in range(len(X_train)):
            test = np.array(X_test.iloc[i,:])
            train = np.array(X_train.iloc[j,:])
            dist = euclidean_dist(train, test)
            distances.append(dist)
        temp = zip(distances, y_train)
        sorted_list = sorted(temp)
        
        labels = []
        for i in range(K):
            labels.append(sorted_list[i][1])
        unique_labels = set(labels)
        labels_res = dict()
        for i in unique_labels:
            labels_res[i] = labels.count(i)
        
        _key = list(labels_res.keys())[0]
        val = list(labels_res.values())[0]
        for key, value in labels_res.items():
            if value > val:
                val = value
                _key = key
        result.append(_key)
        
    return result
```

```python
# l = knn(r2,r1,3)
kk = [2, 9, 18, 21, 23]
res = []
```

```python
for K in kk:
    labels = KNN(K)
    y_test = list(y_test)
    corr = 0
    for i in range(len(labels)):
        if labels[i] == y_test[i]:
            corr += 1
    acc = (corr / len(y_test))*100
    res.append((K, acc))
```
